```python
import pandas as pd
```

```python
artilheiros_br24 = {
    "Yuri Alberto" : ('Corinthians', 15),
    "Alerrandro" : ('Vitória', 15),
    "Estevão" : ('Palmeiras', 13),
    "Vegetti" : ('Vasco da Gama', 12),
    "Pedro" : ('Flamengo', 11),
    "Wesley" : ('Internacional', 11),
    "Raphael Veiga" : ('Palmeiras', 11),
    "Luciano" : ('São Paulo', 11)
}

pd.DataFrame(artilheiros_br24.values(), index=artilheiros_br24.keys(), columns=['Time', 'Gols'])
```

```python

gols = max(artilheiros_br24, key=lambda x: artilheiros_br24[x][1])
time = artilheiros_br24[gols][0]
```

```python
print(f"O artilheiro do Brasileirão 2024 é {gols} do time {time} com {artilheiros_br24[gols][1]} gols.")
print("Parabéns!")
```

```python
for jogador in artilheiros_br24:
    if jogador == gols:
        print(f"Parabéns, {jogador}! Pelo título de artilheiro do Brasileirão 2024.")
    else:
        print(f"Bom trabalho, {jogador}! Continue se esforçando para alcançar o título de artilheiro no próximo ano.")
```

```python
for time in tabela_br24:
    posicao = list(tabela_br24.keys()).index(time) + 1
    pontos = tabela_br24[time][1]
    if posicao <= 6:
        print(f"O time {tabela_br24[time][0].upper()} está na posição {posicao} com {(pontos)} pontos e se classificou para a Copa Libertadores.")
    elif 7 <= posicao <= 8:
        print(f"O time {tabela_br24[time][0].upper()} está na posição {posicao} com {(pontos)} pontos e se classificou para a Pré-Libertadores.")
    elif 9 <= posicao <= 14:
        print(f"O time {tabela_br24[time][0].upper()} está na posição {posicao} com {(pontos)} pontos e se classificou para a Copa Sul-Americana.")
    elif 15 <=posicao >= 16:
        print(f"O time {tabela_br24[time][0].upper()} está na posição {posicao} com {(pontos)} pontos e permaneceu na Série A.")
    else:
        print(f"O time {tabela_br24[time][0].upper()} está na posição {posicao} com {(pontos)} pontos e foi rebaixado para a Série B.")

    
```
